# Step 4: Data Preprocessing

#### Check Categorical Columns

```python
df.select_dtypes(include="object").columns
```

#### Encode Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# Encode all categorical columns
for column in df.select_dtypes(include="object").columns:
    df[column] = le.fit_transform(df[column])
df.head()
```

#### Separate Features and Target

```python
X = df.drop("churn", axis=1)
y = df["churn"]
```

#### Check Shapes

```python
print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

#### Split Training and Testing Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

#### Check Dataset Sizes

```python
print("Training Data :", X_train.shape)
print("Testing Data :", X_test.shape)
```

#### Scale Numerical Features

```python
from sklearn.preprocessing import StandardScaler
# Create scaler object
scaler = StandardScaler()
# Scale training data
X_train = scaler.fit_transform(X_train)
# Scale testing data
X_test = scaler.transform(X_test)
```

```python
type(X_train)
```

#### Verify Final Shapes

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Step 4: Encoded categorical variables, split the dataset into training and testing sets, and standardized numerical features.

# Step 5: Machine Learning Model Building

### Objective
Train multiple machine learning models and compare their performance to find the best model for predicting customer churn.

### Import Required Libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

## Logistic Regression

```python
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
```

```python
lr_prediction = lr_model.predict(X_test)  # Make Predictions
```

```python
lr_accuracy = accuracy_score(y_test, lr_prediction)
print("Logistic Regression Accuracy :", lr_accuracy) # Check Accuracy
```

```python
print(classification_report(y_test, lr_prediction)) # Classification Report
```

```python
print(confusion_matrix(y_test, lr_prediction)) #Confusion Matrix
```

## Decision Tree Classifier

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train) #Train Model
```

```python
dt_prediction = dt_model.predict(X_test) # prediction
```

```python
dt_accuracy = accuracy_score(y_test, dt_prediction)
print("Decision Tree Accuracy :", dt_accuracy) # accuracy
```

```python
print(classification_report(y_test, dt_prediction)) # classification report
```

```python
print(confusion_matrix(y_test, dt_prediction)) # confusion matrix|
```

## Random Forest Classifier

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
#train  model
rf_model.fit(X_train, y_train)
```

```python
rf_prediction = rf_model.predict(X_test)  # prediction
```

```python
rf_accuracy = accuracy_score(y_test, rf_prediction)
print("Random Forest Accuracy :", rf_accuracy) # accuracy
```

```python
print(classification_report(y_test, rf_prediction)) # classication report
```

```python
print(confusion_matrix(y_test, rf_prediction)) # confusion matrix
```

## Compare All Models

```python
result = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "Accuracy": [
        lr_accuracy,
        dt_accuracy,
        rf_accuracy
    ]
})

result
```

## Sort Accuracy

```python
result.sort_values(by="Accuracy", ascending=False)
```

## Accuracy Comparison Chart

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=result,
    x="Model",
    y="Accuracy"
)
plt.title("Model Accuracy Comparison")
plt.show()
```

## Best Model

```python
best_model = result.loc[result["Accuracy"].idxmax()]

print(best_model)
```

## Feature Importance (Random Forest)

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)
importance.head(10)
```

## Visualize Top 10 Features

```python
plt.figure(figsize=(5,4))

sns.barplot(
    data=importance.head(10),
    x="Importance",
    y="Feature"
)

plt.title("Top 10 Important Features")

plt.show()
```

## 2. ROC Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

Logistic Regression ROC

```python
lr_probability = lr_model.predict_proba(X_test)[:,1]
fpr, tpr, threshold = roc_curve(y_test, lr_probability)
plt.figure(figsize=(5,3))
plt.plot(fpr,
         tpr,
         label="Logistic Regression")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

Random Forest ROC

```python
rf_probability = rf_model.predict_proba(X_test)[:,1]
fpr, tpr, threshold = roc_curve(y_test, rf_probability)
plt.figure(figsize=(5,3))
plt.plot(fpr,
         tpr,
         label="Random Forest")

plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
